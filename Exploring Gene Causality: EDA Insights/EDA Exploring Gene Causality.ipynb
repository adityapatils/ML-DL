# 2. Mapping of Phenotypes to Genes

```python
# Install Plotly for interactive visualizations
!pip install plotly

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Step 1: Load the sampled phenotype-gene relationships and causal gene data
phenotype_gene_df = pd.read_csv('/content/sampled_opentargets_step2.for_llm.tsv', sep='\t')
causal_gene_df = pd.read_csv('/content/sampled_opentargets_step2.labels', sep='\t')

# Step 2: Rename columns for clarity
causal_gene_df.columns = ['symbol', 'gene']

# Step 3: Split the ensembl_gene_string into individual genes
phenotype_gene_df['ensembl_gene_list'] = phenotype_gene_df['ensembl_gene_string'].apply(lambda x: x.strip('{}').split('},{'))

# Step 4: Create a new DataFrame to store the results
results = []

# Step 5: Iterate over each row in the phenotype-gene DataFrame
for index, row in phenotype_gene_df.iterrows():
    phenotype = row['description']  # Extract phenotype description
    genes = row['ensembl_gene_list']  # Extract gene list

    # Check which genes are causal
    causal_genes = causal_gene_df[causal_gene_df['gene'].isin(genes)]['gene'].tolist()

    # Label each gene as causal or non-causal
    for gene in genes:
        label = 'Causal' if gene in causal_genes else 'Non-Causal'
        results.append({'Phenotype': phenotype, 'Gene': gene, 'Label': label})

# Step 6: Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Step 7: Display the first few rows of the results DataFrame
print(results_df.head())

# Step 8: Visualization of causal vs non-causal genes

# Count the number of causal and non-causal genes
label_counts = results_df['Label'].value_counts().reset_index()
label_counts.columns = ['Gene Label', 'Count']

# Create an interactive bar plot for causal vs non-causal genes
fig1 = px.bar(
    label_counts,
    x='Gene Label',
    y='Count',
    title='Distribution of Causal vs Non-Causal Genes',
    color='Gene Label',
    color_discrete_sequence=px.colors.qualitative.Set2,
    text='Count'
)
fig1.update_traces(texttemplate='%{text}', textposition='outside')
fig1.update_layout(
    xaxis_title='Gene Label',
    yaxis_title='Count',
    template='plotly_white',
    title_x=0.5
)
fig1.show()


# Count number of genes per phenotype
phenotype_counts = results_df.groupby('Phenotype').agg({
    'Gene': 'count',
    'Label': lambda x: (x == 'Causal').sum()  # Count causal genes
}).reset_index()
phenotype_counts.columns = ['Phenotype', 'Total Genes', 'Causal Genes']

# Calculate Non-Causal Genes
phenotype_counts['Non-Causal Genes'] = phenotype_counts['Total Genes'] - phenotype_counts['Causal Genes']

# Create an interactive scatter plot for causal vs non-causal genes per phenotype
fig2 = px.scatter(
    phenotype_counts,
    x='Non-Causal Genes',
    y='Causal Genes',
    color='Phenotype',
    hover_name='Phenotype',
    size='Total Genes',  # Size based on total genes for added dimension
    title='Causal vs Non-Causal Genes per Phenotype',
    labels={'Non-Causal Genes': 'Number of Non-Causal Genes', 'Causal Genes': 'Number of Causal Genes'},
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig2.update_layout(
    xaxis_title='Non-Causal Genes',
    yaxis_title='Causal Genes',
    template='plotly_white',
    title_x=0.5
)
fig2.show()

# Print summary statistics
print(f"\nTotal number of phenotypes analyzed: {len(phenotype_counts)}")
print(f"Total genes analyzed: {results_df['Gene'].nunique()}")
```

# 4. Vector Analysis

```python
# Install necessary libraries if not already installed
!pip install plotly

import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine
from sklearn.decomposition import PCA
import plotly.express as px
import plotly.graph_objects as go
from ast import literal_eval

# Load the datasets using the 'python' engine and error handling
gene_df = pd.read_csv('/content/sampled_gene_embeddings.csv', engine='python', on_bad_lines='skip')  # Robust parsing
phenotype_df = pd.read_csv('/content/sampled_phenotype_embeddings.csv')

# Convert embedding strings to numpy arrays
gene_df['embedding'] = gene_df['embedding'].apply(literal_eval)
phenotype_df['embedding'] = phenotype_df['embedding'].apply(literal_eval)

# Function to calculate cosine similarity
def cosine_similarity(v1, v2):
    return 1 - cosine(v1, v2)

# Convert embeddings to numpy arrays
phenotype_vectors = np.array(phenotype_df['embedding'].tolist())
gene_vectors = np.array(gene_df['embedding'].tolist())

# Assuming a one-to-one mapping for simplicity in this example
num_samples = min(len(phenotype_vectors), len(gene_vectors))
derived_vectors = phenotype_vectors[:num_samples] - gene_vectors[:num_samples]

# Calculate cosine similarity between phenotype and gene vectors
cosine_similarities = [cosine_similarity(phenotype_vectors[i], gene_vectors[i]) for i in range(num_samples)]

# Displaying important metrics
print(f"Number of samples processed: {num_samples}")
print(f"Shape of derived vectors: {derived_vectors.shape}")
print(f"Number of cosine similarities calculated: {len(cosine_similarities)}")
print(f"Range of cosine similarities: {min(cosine_similarities):.4f} to {max(cosine_similarities):.4f}")

# Display the first few cosine similarities
print("First few cosine similarities:")
print(cosine_similarities[:5])

# Visualization 1: Interactive Histogram of Cosine Similarities
fig1 = px.histogram(
    x=cosine_similarities,
    title='Cosine Similarity between Phenotype and Gene Vectors',
    labels={'x': 'Cosine Similarity'},
    nbins=30,
    marginal='box'
)
fig1.show()

# PCA for dimensionality reduction of derived vectors
pca_derived = PCA(n_components=2).fit_transform(derived_vectors)

# Create a DataFrame for PCA results
pca_df = pd.DataFrame(pca_derived, columns=['PC1', 'PC2'])
pca_df['Cosine Similarity'] = cosine_similarities[:num_samples]

# Visualization 2: Interactive Scatter Plot of PCA Results
fig2 = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    color='Cosine Similarity',
    title='PCA of Derived Vectors (Phenotype - Gene)',
    labels={'PC1': 'First Principal Component', 'PC2': 'Second Principal Component'},
    color_continuous_scale=px.colors.sequential.Viridis,
    hover_name=pca_df.index
)
fig2.show()

# Print detailed insights
print("\nDetailed Insights:")
print("---------------------------------------------------")
print(f"Total Samples Processed: {num_samples}")
print(f"Max Cosine Similarity: {max(cosine_similarities):.4f} at index {cosine_similarities.index(max(cosine_similarities))}")
print(f"Min Cosine Similarity: {min(cosine_similarities):.4f} at index {cosine_similarities.index(min(cosine_similarities))}")
print(f"Mean Cosine Similarity: {np.mean(cosine_similarities):.4f}")
```

# 5. Clustering

```python
# Install necessary libraries if not already installed
!pip install plotly

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
import plotly.express as px
from ast import literal_eval

# Load the datasets using the 'python' engine and error handling
gene_df = pd.read_csv('/content/sampled_gene_embeddings.csv', engine='python', on_bad_lines='skip')  # Robust parsing
phenotype_df = pd.read_csv('/content/sampled_phenotype_embeddings.csv')

# Convert embedding strings to numpy arrays
gene_df['embedding'] = gene_df['embedding'].apply(literal_eval)
phenotype_df['embedding'] = phenotype_df['embedding'].apply(literal_eval)

# Combine the embeddings for clustering
combined_embeddings = np.vstack((gene_df['embedding'].tolist(), phenotype_df['embedding'].tolist()))

# Standardize the data
scaler = StandardScaler()
scaled_embeddings = scaler.fit_transform(combined_embeddings)

# K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_embeddings)

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_embeddings)

# Perform PCA for visualization
pca = PCA(n_components=2)
pca_embeddings = pca.fit_transform(scaled_embeddings)

# Prepare DataFrame for visualization
clustered_df = pd.DataFrame(pca_embeddings, columns=['PC1', 'PC2'])
clustered_df['KMeans Cluster'] = kmeans_labels
clustered_df['DBSCAN Cluster'] = dbscan_labels

# Visualization 1: Interactive K-means Clustering Plot
fig1 = px.scatter(
    clustered_df,
    x='PC1',
    y='PC2',
    color='KMeans Cluster',
    title='K-means Clustering of Gene and Phenotype Embeddings',
    labels={'PC1': 'First Principal Component', 'PC2': 'Second Principal Component'},
    color_discrete_sequence=px.colors.qualitative.Plotly,  # Use distinct colors
    hover_data={'PC1': True, 'PC2': True, 'KMeans Cluster': True, 'DBSCAN Cluster': True}
)
fig1.show()

# Visualization 2: Interactive DBSCAN Clustering Plot with Enhanced Hover Information
fig2 = px.scatter(
    clustered_df,
    x='PC1',
    y='PC2',
    color='DBSCAN Cluster',
    title='DBSCAN Clustering of Gene and Phenotype Embeddings',
    labels={'PC1': 'First Principal Component', 'PC2': 'Second Principal Component'},
    color_discrete_sequence=px.colors.qualitative.Plotly,  # Use distinct colors
    hover_data={
        'PC1': True,
        'PC2': True,
        'KMeans Cluster': True,
        'DBSCAN Cluster': True,
    },
)
fig2.update_traces(hovertemplate="<br>".join([
    "First Principal Component: %{x:.2f}",
    "Second Principal Component: %{y:.2f}",
    "KMeans Cluster: %{customdata[0]}",
    "DBSCAN Cluster: %{customdata[1]}"
]),
    customdata=np.stack((clustered_df['KMeans Cluster'], clustered_df['DBSCAN Cluster']), axis=-1)
)

fig2.show()

# Print detailed insights
print("\nDetailed Insights:")
print("---------------------------------------------------")
print(f"Number of samples processed: {len(clustered_df)}")
print(f"KMeans Unique Clusters: {len(clustered_df['KMeans Cluster'].unique())}")
print(f"DBSCAN Unique Clusters: {len(clustered_df['DBSCAN Cluster'].unique())}")
print(f"KMeans Cluster Sizes:\n{clustered_df['KMeans Cluster'].value_counts()}")
print(f"DBSCAN Cluster Sizes:\n{clustered_df['DBSCAN Cluster'].value_counts()}")
```

#Comparison of performing the tasks using the **full dataset** versus a **sample subset dataset** (created smaple dataset):

| **Criteria**                        | **Full Dataset**                                             | **Sample Subset Dataset**                                      |
|-------------------------------------|--------------------------------------------------------------|----------------------------------------------------------------|
| **Size**                            | Large, contains all phenotypes and genes.                    | Small, contains a randomly sampled 500 phenotypes.             |
| **Computational Efficiency**        | Slower due to larger data, requires more memory and time.     | Faster processing, suitable for limited computational resources.|
| **Analysis Focus**                  | Broad, covers all available data, ideal for global insights.  | Narrowed focus on selected phenotypes for specific analysis.    |
| **Reproducibility**                 | Harder to reproduce due to full-size complexity.              | Easily reproducible due to hash-based random sampling.          |
| **Dimensionality Reduction**        | High complexity, harder to visualize, but more comprehensive. | Simpler, faster to reduce, easier visualization with smaller data.|
| **Vector Analysis**                 | Detailed, captures wide relationships, but computationally expensive.| Quicker, allows focused exploration of key relationships.       |
| **Clustering**                      | Complex clustering, potentially revealing more hidden patterns.| Easier to cluster, quicker insights into relationships.         |
| **Validation**                      | Strong, covers all data, provides global context for causality.| May miss out on rare or unique cases present only in full data. |

# In summary, use **full dataset** for comprehensive global insights and robustness, while the **sample subset dataset** is ideal for focused, efficient, and reproducible analysis.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression

# Load the dataset with better error handling
try:
    df = pd.read_csv('/content/opentargets_step2.for_llm.tsv', sep='\t', error_bad_lines=False)
    print("Dataset loaded successfully.")
except Exception as e:
    print(f"Error loading dataset: {e}")

# Check the first few rows and the column names
print(df.head())
print(df.columns)

# Function to count genes for each phenotype safely
def count_genes(gene_string):
    try:
        return len(gene_string.strip('{}').split('},{'))
    except Exception as e:
        print(f"Error processing gene string: {e}")
        return 0

# Add a column with the count of genes for each phenotype
if 'symbol_gene_string' in df.columns:
    df['gene_count'] = df['symbol_gene_string'].apply(count_genes)

    # Calculate some statistics
    total_phenotypes = len(df)
    total_genes = df['gene_count'].sum()
    avg_genes_per_phenotype = df['gene_count'].mean()
    max_genes = df['gene_count'].max()
    min_genes = df['gene_count'].min()

    print(f"Total phenotypes: {total_phenotypes}")
    print(f"Total genes: {total_genes}")
    print(f"Average genes per phenotype: {avg_genes_per_phenotype:.2f}")
    print(f"Max genes for a phenotype: {max_genes}")
    print(f"Min genes for a phenotype: {min_genes}")

    # Create interactive histogram
    fig_hist = px.histogram(df, x='gene_count', nbins=30, title='Distribution of Gene Counts per Phenotype',
                             labels={'gene_count': 'Number of Genes'},
                             histnorm='probability density')
    fig_hist.update_traces(opacity=0.75)
    fig_hist.show()

    # Create interactive box plot
    fig_box = px.box(df, y='gene_count', title='Box Plot of Gene Counts per Phenotype',
                     labels={'gene_count': 'Number of Genes'})
    fig_box.show()

    # Get the top 10 phenotypes with the most genes
    top_10_phenotypes = df.nlargest(10, 'gene_count')

    # Create interactive bar plot of the top 10 phenotypes
    fig_bar = px.bar(top_10_phenotypes, x='description', y='gene_count',
                     title='Top 10 Phenotypes with the Most Associated Genes',
                     labels={'description': 'Phenotype', 'gene_count': 'Number of Genes'},
                     color='gene_count',
                     text='gene_count')
    fig_bar.update_traces(texttemplate='%{text:.2f}', textposition='outside')
    fig_bar.update_layout(xaxis_tickangle=-45)
    fig_bar.show()

    # Linear Regression Model
    df['description_length'] = df['description'].str.len()

    X = df[['description_length']]
    y = df['gene_count']

    model = LinearRegression()
    model.fit(X, y)

    df['predicted_gene_count'] = model.predict(X)

    print(f"Intercept: {model.intercept_}, Coefficient: {model.coef_[0]}")

    # Interactive Plotting of Actual vs Predicted Gene Counts
    fig_pred = go.Figure()
    fig_pred.add_trace(go.Scatter(x=df['description_length'], y=df['gene_count'],
                                    mode='markers', name='Actual Gene Count', marker=dict(color='blue')))
    fig_pred.add_trace(go.Scatter(x=df['description_length'], y=df['predicted_gene_count'],
                                    mode='markers', name='Predicted Gene Count', marker=dict(color='red', opacity=0.5)))

    fig_pred.update_layout(title='Actual vs Predicted Gene Counts',
                           xaxis_title='Phenotype Description Length',
                           yaxis_title='Gene Count',
                           showlegend=True)
    fig_pred.show()

    print("\nFirst few rows of the dataset:")
    print(df.head().to_string())
else:
    print("Column 'symbol_gene_string' not found in the DataFrame.")
```

# 7.Vector Analysis Larger Dataset

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import plotly.express as px
import json  # Import JSON module for safe evaluation

# Load the sampled embeddings with additional error handling
try:
    # Use Python engine and skip bad lines to handle parsing issues
    sampled_gene_embeddings = pd.read_csv('/content/gene_embeddings.csv', engine='python', on_bad_lines='skip')
    sampled_phenotype_embeddings = pd.read_csv('/content/phenotype_embeddings.csv', engine='python', on_bad_lines='skip')
    print("Files loaded successfully.")

except pd.errors.ParserError as e:
    print(f"Parsing error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")

# Safe evaluation function for embeddings
def safe_eval(embedding_string):
    try:
        return json.loads(embedding_string)
    except json.JSONDecodeError:
        print(f"Warning: Failed to decode embedding: {embedding_string}")
        return None  # Handle as appropriate

# Extract the embeddings using the safe_eval function
gene_embeddings = sampled_gene_embeddings['embedding'].apply(safe_eval).dropna().tolist()
phenotype_embeddings = sampled_phenotype_embeddings['embedding'].apply(safe_eval).dropna().tolist()

# Convert to numpy arrays
gene_embeddings = np.array(gene_embeddings)
phenotype_embeddings = np.array(phenotype_embeddings)

# Step 1: Ensure both embeddings have the same number of samples for subtraction
min_samples = min(len(gene_embeddings), len(phenotype_embeddings))

# Step 2: Subset the embeddings to the minimum number of samples
gene_embeddings_subset = gene_embeddings[:min_samples]
phenotype_embeddings_subset = phenotype_embeddings[:min_samples]

# Step 3: Perform vector subtraction: phenotype_embeddings - gene_embeddings
derived_vectors = phenotype_embeddings_subset - gene_embeddings_subset

# Step 4: Apply PCA to reduce dimensions to 2D for visualization
pca_derived = PCA(n_components=2)
derived_vectors_2d = pca_derived.fit_transform(derived_vectors)

# Step 5: Create a DataFrame for visualization
pca_df = pd.DataFrame(derived_vectors_2d, columns=['PC1', 'PC2'])

# Add additional information if necessary
pca_df['Label'] = ['Sample ' + str(i) for i in range(len(pca_df))]  # Placeholder labels

# Step 6: Create an interactive scatter plot using Plotly
fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    title='2D PCA of Derived Vectors (Phenotype - Gene)',
    hover_name='Label',  # Change this to your actual labels if available
    color_discrete_sequence=['purple'],
    opacity=0.7
)

# Customize layout for better visualization
fig.update_traces(marker=dict(size=8))
fig.update_layout(
    xaxis_title='Principal Component 1',
    yaxis_title='Principal Component 2',
    template='plotly_white',
    title_x=0.5
)

# Show the interactive plot
fig.show()
```

# Clustering Larger Dataset

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import plotly.express as px
from scipy.spatial.distance import cdist

# Load the sampled embeddings with error handling
try:
    gene_df = pd.read_csv('/content/gene_embeddings.csv', engine='python', on_bad_lines='skip')
    phenotype_df = pd.read_csv('/content/phenotype_embeddings.csv', engine='python', on_bad_lines='skip')

    print("Files loaded successfully.")

except pd.errors.ParserError as e:
    print(f"Parsing error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")

# Extract embeddings
gene_embeddings = np.array(gene_df['embedding'].apply(eval).tolist())
phenotype_embeddings = np.array(phenotype_df['embedding'].apply(eval).tolist())

# Function to perform clustering and interactive visualization
def cluster_and_visualize(embeddings, title, n_clusters=5):
    # Perform K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(embeddings)

    # Reduce dimensionality for visualization
    pca = PCA(n_components=2)
    reduced_embeddings = pca.fit_transform(embeddings)

    # Create a DataFrame for Plotly
    pca_df = pd.DataFrame(reduced_embeddings, columns=['PC1', 'PC2'])
    pca_df['Cluster'] = cluster_labels

    # Create an interactive scatter plot using Plotly
    fig = px.scatter(
        pca_df,
        x='PC1',
        y='PC2',
        color='Cluster',
        title=f'{title} - K-Means Clustering (k={n_clusters})',
        color_continuous_scale=px.colors.sequential.Viridis,
        labels={'PC1': 'First Principal Component', 'PC2': 'Second Principal Component'},
        hover_data={'Cluster': True}
    )

    # Customize layout
    fig.update_traces(marker=dict(size=8, opacity=0.7))
    fig.update_layout(
        template='plotly_white',
        title_x=0.5
    )

    fig.show()

    return cluster_labels

# Cluster and visualize gene embeddings
gene_clusters = cluster_and_visualize(gene_embeddings, 'Gene Embeddings')

# Cluster and visualize phenotype embeddings
phenotype_clusters = cluster_and_visualize(phenotype_embeddings, 'Phenotype Embeddings')

# Analyze cluster distributions
gene_cluster_dist = pd.Series(gene_clusters).value_counts().sort_index()
phenotype_cluster_dist = pd.Series(phenotype_clusters).value_counts().sort_index()

# Visualize cluster distributions with Plotly
def plot_cluster_distribution(cluster_dist, title):
    fig = px.bar(
        x=cluster_dist.index,
        y=cluster_dist.values,
        labels={'x': 'Cluster', 'y': 'Count'},
        title=title,
        color=cluster_dist.values,
        color_continuous_scale='Blues'
    )

    fig.update_layout(
        template='plotly_white',
        title_x=0.5
    )

    fig.show()

# Plot distributions
plot_cluster_distribution(gene_cluster_dist, 'Gene Cluster Distribution')
plot_cluster_distribution(phenotype_cluster_dist, 'Phenotype Cluster Distribution')

# Calculate cluster centroids
kmeans_genes = KMeans(n_clusters=5, random_state=42).fit(gene_embeddings)
kmeans_phenotypes = KMeans(n_clusters=5, random_state=42).fit(phenotype_embeddings)

gene_centroids = kmeans_genes.cluster_centers_
phenotype_centroids = kmeans_phenotypes.cluster_centers_

# Calculate distances between gene and phenotype centroids
centroid_distances = cdist(gene_centroids, phenotype_centroids)

# Visualize centroid distances using Plotly
distances_df = pd.DataFrame(centroid_distances, columns=[f'Phenotype Cluster {i+1}' for i in range(phenotype_centroids.shape[0])])
distances_df.index = [f'Gene Cluster {i+1}' for i in range(gene_centroids.shape[0])]

fig = px.imshow(
    distances_df,
    title='Distances Between Gene and Phenotype Cluster Centroids',
    labels=dict(x='Phenotype Clusters', y='Gene Clusters'),
    color_continuous_scale='YlGnBu'
)

fig.update_layout(title_x=0.5)
fig.show()
```
